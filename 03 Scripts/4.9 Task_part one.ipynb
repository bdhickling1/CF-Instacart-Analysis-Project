## 3.1 Wrangle data for logic consistency

```python
# Drop columns that are not required for the analysis (First Name, Surnam)
df_customers = df_customers.drop(columns = ['First Name', 'Surnam'])
```

```python
# Verify dropped columns
df_customers.head()
```

```python
# Rename columns for consistent logic (Gender - gender, STATE - state, Age - age, n_dependents - family_dependents, fam_status - family_status)
df_customers.rename(columns = {'Gender' : 'gender', 'STATE' : 'state', 'Age' : 'age', 'n_dependants' : 'family_dependants', 'fam_status' : 'family_status'}, inplace = True)
```

```python
# Verify renamed columns
df_customers.head()
```

```python
# Check columns for missing data
df_customers.isnull().sum()
```

__There is no missing data in the dataset__

```python
# Check for any data types that need to be changed
df_customers.info()
```

__'user_id' is listed as an integer, but should be a string. 'age', 'family_dependants", and 'income' should all stay as integers as they would be useful in a statistical analysis. All columns listed as object, including 'date_joined' should remain as objects__

```python
# Change data type of 'user_id'
df_customers['user_id'] = df_customers['user_id'].astype('str')
```

```python
# Check change of data type
df_customers['user_id'].dtype
```
