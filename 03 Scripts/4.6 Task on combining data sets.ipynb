###### __As the columns and rows between the two dataframes are not the same length, nor are there the same # of columns, the merge function is the most appropriate. There is a matching column between the two DataFrames, 'product_id', on which to conduct a merge. As I expect a fully matching 'product_id' column, I don't need to specify a join as an inner join will be used by default__

```python
# Merge the df_ords_prior_prods and df_prods dataframes
df_merged = df_prods.merge(df_ords_prior_prods, on = 'product_id', indicator = True)
```

____I researched the use of the '_merge' column using forums on Stack Overflow. It appears that if you try to merge two dataframes and one of the dataframes have alredy been merged you will get an error as a second '_merge' column will be created that will cause the error.  In this case it was suggested to use the argument 'indicator  = 'exists', in place of 'indicator = True'. In my case, as I dropped the '_merge' column from the first merged dataframe after importing it into this task, this wasn't necessary do do. It was a good lesson to learn for performing future merges.__
