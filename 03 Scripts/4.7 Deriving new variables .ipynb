# 03.1 Create a price range flag using the if-statement function user-defined function

```python
# Limit dataframe to first 1,000,000 rows by creating a subset
df = df_ords_prods_merge[:1000000]
```

```python
df
```

```python
df.shape
```

```python
# Create a 'price-range' flag using the 'if-else' function
# Define the function and create conditions for the flag
def price_lable(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif (row['prices'] > 5) and (row['prices'] <= 15):
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else: return np.nan
```

```python
# Apply the flag on the dataframe
df['price_range'] = df.apply(price_lable, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Check for the most expensive product within the subset
df['prices'].max()
```

# 03.3 Use if-statement with for-loop to identify the busiest and slowest day of the week for customer orders

```python
# Print the frequency of the 'orders_day_of_the_week' column to find out which days most orders take place
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
# Create a new column for 'busiest day' with flags for busiest (0), regularly busy (others), and least busy (4) days
result = []

for value in df_ords_prods_merge['orders_day_of_week']:
    if value == 0:
        result.append('Busiest day')
    elif value == 4:
        result.append('Least busy')
    else:
        result.append('Regularly busy')
```

```python
result
```

```python
# Create a new column in the ords_prods_merge dataframe using the result (flag of business of each day for orders)
df_ords_prods_merge['busiest_day'] = result
```

```python
# Check results by printing the frequency of 'busiest_days' and compare with 'orders_day_of_week'
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

# 04.3 Task 4.7 Step 4 - to address the situation when two many users make orders at the same time, which can cause the Instacart app to freeze, identify, for the Senior Technical Officer, the busiest hours of the day using 'most orders', 'average orders', and 'fewest orders' as the 'busiest_period_of_day'.

```python
# Print the frequency of each hour of the day to identify the busiest and least busiest hours for orders
df_ords_prods_merge['order_hour_of_day'].value_counts()
```

```python
# Identify the average hour for orders
df_ords_prods_merge['order_hour_of_day'].max()
```

__As the distribution is skewed right, you can't take the mean as average. To group the 'order_hour_of_day' by most, least and average it is advised to use the 75th, 25th and 50th quantiles__

```python
# Create a list showing the quantiles for the frequency of order_hour_of_day values
list(df_ords_prods_merge['order_hour_of_day'].value_counts().quantile([0.25, 0.5, 0.75]))
```

__Looking at the frequency and quantile distribution, you can order the hours by most (10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18, 19) and least (3, 4, 2, 5, 1, 0).__

```python
# Create a for-loop for the most, average, and fewest order periods of the day using the 75, 25 and 
result3 = []

for value in df_ords_prods_merge['order_hour_of_day']:
    if value in [10, 11, 14, 15, 13, 12, 16, 9, 17, 8, 18, 19]:
        result3.append('Most orders')
    elif value in [3, 4, 2, 5, 1, 0]:
        result3.append('Fewest orders')
    else:
        result3.append('Average orders')
```

```python
df_ords_prods_merge['busiest_period_of_day'] = result3
```
