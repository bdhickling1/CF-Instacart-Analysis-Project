# 04. Group data with pandas

```python
# Step 1 of grouping - split data into groups by criteria
df.groupby('product_name')
```

__Groupby() has not output on its own - it creats an object, product_name in this case. You need to apply a function or aggregate data, etc. to achieve an output__

# 05. Aggregating data with agg()

```python
# Group data by 'department_id' and aggregate the 'order_number' data to get the average of orders by department group
df.groupby('department_id').agg({'order_number': ['mean']})
```

```python
# Aggregate without using agg() - obtain mean of 'order_number' using .mean() function
df.groupby('department_id')['order_number'].mean()
```

# 06. Aggregating data with transform()

```python
# Create flags with aggregated data - create a loyalty flag
# Use transform() to create a new column containing the maximum frequency of the 'order_number' column, then loc() to create a second column with a flag designating whether a customer is loyal or not.
# Customers with orders > 40 = loyal. Customers with 10 - 40 are regular, and <= 10 are new.
# Step 1-3 for max orders - split the data into groups
df['max_order'] = df.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Deriving columns with loc()
df.loc[df['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
df.loc[(df['max_order'] <=40) & (df['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
df.loc[df['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Print the frequency of the 'loyalty_flag' column
df['loyalty_flag'].value_counts(dropna = False)
```
